```python
import pandas as pd
from datetime import datetime

# Ensure BirthDate is converted properly
df1['BirthDate'] = pd.to_datetime(df1['BirthDate'], errors='coerce')

# Check if conversion worked
print(df1['BirthDate'].head())

# Drop rows where conversion failed (optional)
df1 = df1.dropna(subset=['BirthDate'])

# Calculate age
today = pd.Timestamp.today()
df1['Age'] = (today - df1['BirthDate']).dt.days // 365

print(df1[['BirthDate', 'Age']].head())
```

```python
bins = [0, 18, 50, df1['Age'].max() ]
GroupNames = ['Young', 'Adult', 'Senior']
df1['AgeBins'] =pd.cut(df1['Age'], bins, labels=GroupNames, include_lowest=True)
df1[['Age', 'AgeBins']].head()
```

# PART 3

### a)

```python
obj1 = df_encoded.iloc[0].values
obj2 = df_encoded.iloc[1].values

# --- Simple Matching Coefficient (SMC) ---
matches = np.sum(obj1 == obj2)
smc = matches / len(obj1)
print(smc)
```

```python
# --- Jaccard Similarity ---
intersection = np.sum((obj1 == 1) & (obj2 == 1))
union = np.sum((obj1 == 1) | (obj2 == 1))
jaccard = intersection / union if union != 0 else 0
print(jaccard)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# 1. Drop useless columns
df_clean = df_encoded.drop(columns=['FirstName', 'LastName', 'AddressLine1', 
                            'City', 'StateProvinceName', 
                            'CountryRegionName', 'PostalCode', 
                            'BirthDate'])

# 2. Separate categorical and numeric
categorical_cols = ['Education', 'Occupation', 'AgeBins', 'IncomeCat']
numeric_cols = ['HomeOwnerFlag', 'NumberCarsOwned', 'NumberChildrenAtHome', 
                'TotalChildren', 'YearlyIncome', 'Age']

# 3. One-Hot Encode categorical columns
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded = encoder.fit_transform(df_clean[categorical_cols])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_cols))

# 4. Merge back with numeric features
df_ready = pd.concat([df_clean[numeric_cols].reset_index(drop=True), encoded_df], axis=1)

```

```python
#--- Cosine Similarity ---
obj1 = df_ready.iloc[0].values
obj2 = df_ready.iloc[1].values
from sklearn.metrics.pairwise import cosine_similarity
cosine=cosine_similarity([obj1],[obj2])[0][0]
print(cosine)
```
