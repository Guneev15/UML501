```python
import cv2
import numpy as np
import os
def img_to_array(path):
    img = cv2.imread(path)

    if img is None:
        print("Image not found or path is incorrect.")
        return

    if len(img.shape) == 2:
        print("Detected grayscale image.")
        filename = 'grayscale_image.txt'
        np.savetxt(filename, img, fmt='%d')
        
    elif len(img.shape) == 3:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        print("Detected RGB image.")
        for i, color in enumerate(['R', 'G', 'B']):
            np.savetxt(f'{color}_channel.txt', img_rgb[:, :, i], fmt='%d')
    else:
        print("Unknown image format.")
        
img_to_array("C:/Users/Gunee/OneDrive/Desktop/website.jpg")
```
