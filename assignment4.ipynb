# Q1

```python
from bs4 import BeautifulSoup
import requests
url="https://books.toscrape.com/"
page=requests.get(url)
soup=BeautifulSoup(page.content,"html.parser")
```

### 1)Title

```python
books=soup.find_all("article",class_="product_pod")
for book in books:
    title=book.h3.a["title"]
    print(title)
```

### 2) Price

```python
for book in books:
    title=book.h3.a["title"]
    price=book.find("p",class_="price_color").text
    print(title," -> ",price)
```

### 3)

```python
for book in books:
    title=book.h3.a["title"]
    stock=book.find("p",class_="instock availability").text.strip()
    print(title,"->",stock)
```

### 4)

```python
i=1
data=[]
for book in books:
    title=book.h3.a["title"]
    price=book.find("p",class_="price_color").text
    stock=book.find("p",class_="instock availability").get_text(strip=True)
    rating_tag=book.find("p",class_="star-rating")
    ratings=rating_tag["class"][1]
    print(f" {i}) {title} → {price} → {stock} → {ratings} Stars")
    i=i+1
    data.append([title,price,stock,ratings])
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=["Title","Price","Stock","Ratings"])
df.to_csv("books.csv",index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver=webdriver.Chrome()
driver.get("https://www.imdb.com/chart/top/")
```

```python
movies=driver.find_elements(By.CSS_SELECTOR,"ul.ipc-metadata-list ipc-metadata-list--dividers-between sc-2b8fdbce-0 emPbxy compact-list-view ipc-metadata-list--base")
```

```python
data=[]
for movie in movies[:10]:
    title_block = movie.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text
    rank, title = title_block.split(". ", 1)
    year = movie.find_element(By.CSS_SELECTOR, "span.sc-15ac7568-7.cCsint.cli-title-metadata-item").text
    year = year.strip("()")
    rating = movie.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating").text
    print(rank,title,"->",year,"->",rating)
    data.append([int(rank), title, int(year), float(rating)])
```

```python
driver.quit()
```

```python
df = pd.DataFrame(data, columns=["Rank", "Title", "Year", "IMDB Rating"])
df.to_csv("imdb_top250.csv", index=False)
df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the weather page
url = "https://www.timeanddate.com/weather/"

# Send request to the website
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Find the weather table for top cities
table = soup.find("table", {"class": "zebra.tb-wt.fw.va-m"})

cities = []
temperatures = []
conditions = []

# Extract rows (skip header row)
for row in table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) >= 3:
        city = cols[0].get_text(strip=True)
        temp = cols[1].get_text(strip=True)
        condition = cols[2].get_text(strip=True)

        cities.append(city)
        temperatures.append(temp)
        conditions.append(condition)

# Create a DataFrame
weather_df = pd.DataFrame({
    "City": cities,
    "Temperature": temperatures,
    "Condition": conditions
})

# Save to CSV
weather_df.to_csv("weather.csv", index=False)

print("Scraping completed! Data saved to weather.csv")
print(weather_df.head())
```
